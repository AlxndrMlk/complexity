## Dissipation

Dissipation happens when state space shrinks.


**Liouville's theorem** states that dissipation is related to state space shrinkage.

## Lyapunov exponents


Let's denote **Lyapunov exponent** as $\lambda$.


* $n$-dimensional system has $n$ $\lambda_i$s


* negative $\lambda_i$ compresses state space along *stable manifold* (initial conditions converge to a fixed-point)


* positive $\lambda_i$ stretches it along *unstable manifold* (initial conditions diverge into infinity)

__________


* For **dissipative systems**:


$$\large \sum_{i=0}^{n} \lambda_i < 0 $$


* For **chaotic systems**


$$\large \exists_{\lambda_i} \  \lambda_i > 0 $$

    (but it's not NASC for chaos)
    
____________

**Lyapunov exponents** are properties of attrectors. They hold true for all initial conditions within the basin of attraction for a given attractor.

```python
x = np.arange(-5, 5.1, .1)
y = np.e**x
```

```python
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$e^x$')
plt.title('$x$ vs $e^x$')
plt.show()
```

**NOTE**: The largest **Lyapunov exponent** denoted $\lambda_1$, dominates, when $t \rightarrow \infty$.
