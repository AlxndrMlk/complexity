## Maps



Maps are discrete-time mappings of value at $t-1$ to the value at $t$. They are formalized using difference equations.


I.e.:


$$\large x_{n + 1} = f(x_n) $$


It could be for instance:


$$\large x_{n + 1} = cos(x_n) $$

```python
def cos_map(x, n_steps = 1):
    
    x_ = np.radians(x)
    process = [x_]
    
    for i in range(n_steps):
        x_ = np.cos(x_)
        process.append(x_)
        
    return np.array(process)
```

```python
vals_90  = cos_map(90, 18)
vals_5  = cos_map(5, 18)
vals_48 = cos_map(48, 18)
```

```python
plt.figure(figsize = (12, 4))
plt.plot(vals_90, alpha = .5, label = 'degrees = 90')
plt.plot(vals_5, alpha = .5, label = 'degrees = 5')
plt.plot(vals_48, alpha = .5, label = 'degrees = 48')
plt.title('Cosine map')
plt.legend()
plt.show()
```

As we can see, cosine map converges to a point attractor.

Another map is a ogistic map:


$$\large x_{n+1} = R x_n (1 - x_n)$$

```python
def logistic_map(x, r, n_steps = 1):
    
    x_ = x
    process = [x_]
    
    for i in range(n_steps):
        x_ = r * x_ * (1 - x_)
        process.append(x_)
        
    return np.array(process)
```

```python
log_map_2 = logistic_map(.2, 2, 18)
log_map_5 = logistic_map(.2, 4, 18)
log_map_7 = logistic_map(.2, 3.3, 18)
```

```python
plt.figure(figsize = (12, 4))
plt.plot(log_map_2, alpha = .5, label = '$x_0 = .2; R = 2$')
plt.plot(log_map_5, alpha = .5, label = '$x_0 = .2; R = 4$')
plt.plot(log_map_7, alpha = .5, label = '$x_0 = .2; R = 3.3$')
plt.title('Logistic map')
plt.legend()
plt.show()
```

Use your logistic map program to explore the idea of sensitive dependence on initial conditions.  Generate two trajectories of the logistic map $\{x\}$ and $\{\hat{x}\}$.  For each of these experiments the initial condition of $\{x\}$ should be $x_0 = 0.2$ and for $\{\hat{x}\}$ you should use $x_0 = 0.200001$.  For now, fix $r = 2$, generate these two trajectories and then plot $|x_n - \hat{x_n}|$ vs $n$.  Now repeat this for $r = 3.4$, and $r = 3.72$.  Compare your results to the three plots below and then answer the questions that follow.

```python
x0_1 = .2
x0_2 = .200001

n = 200
```

```python
def plot_exp(xs, r, n):
    
    processes = []
    
    for x in xs:
        processes.append(logistic_map(x, r, n))

    plt.plot(np.abs(processes[0] - processes[1]), 
             alpha = .3, 
             linewidth = 2,
             color = 'gray', 
             marker = 'o', 
             markerfacecolor = 'darkblue', 
             label = f'$R = {r}$')
    plt.legend()
    plt.show()
```

```python
for r_val in [2, 3.4, 3.72]:
    plot_exp([x0_1, x0_2], r_val, n)
```
