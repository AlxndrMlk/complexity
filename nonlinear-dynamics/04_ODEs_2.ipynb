## Taylor series

**Taylor series** is a mathematical recipe for approximating a given function $f(x)$ near some point $a$ using polynomials.

A general form of **Talyor series** is:


$$ \large P_n(x) = f(a) + (x - a)f'(x) + \frac{(x - a)^2}{2!}f''(a) + \ ... \ + \frac{(x - a)^n}{n!}f^{(n)}(a)$$

$$ \large P_n(x) = f(a) + \sum_{n=1}^{n + 1}\frac{(x - a)^n}{n!}f^{(n)}(a)$$

```python
# Approx e^x near a = 0 by x + 1
x = np.arange(-1, 1.1, .1)
```

```python
plt.figure(figsize = (12, 6))
plt.title('Taylor series approx. of $e^x$ near $a = 0$')
plt.plot(x, np.e**x, label = '$f(x) = e^x$', alpha = .5)
plt.plot(x, x + 1, label = '$f(x) = x + 1$', alpha = .5)
plt.plot(x, 1 + x + x**2 / 2, label = '$f(x) = 1 + x + x^2 / 2$', alpha = .5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Shadowing lemma

Noise-added trajectory on chaotic attractor is *shadowed* by a true trajectory.

This is only true as long as the noise doesn't bump trajectory out of the basin of attraction.

It also hold true only for state noise - **not parameter noise**.
